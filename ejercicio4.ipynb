```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
```

```python
img_coca = io.imread('coca-cola.jpeg')
img_pepsi = io.imread('pepsi.png')

# Convertir a HSV
hsv_coca = color.rgb2hsv(img_coca)
hsv_pepsi = color.rgb2hsv(img_pepsi)

# Extraer el canal de matiz (Hue)
hue_coca = hsv_coca[:, :, 0]
hue_pepsi = hsv_pepsi[:, :, 0]

def plot_histogram(hue_channel, title, color):
    plt.hist(hue_channel.ravel(), bins=256, color=color, alpha=0.7)
    plt.title(title)
    plt.xlabel("Valor de Hue")
    plt.ylabel("Frecuencia")
```

```python

# Graficar histogramas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_histogram(hue_coca, "Histograma de Hue - Coca Cola", "red")

plt.subplot(1, 2, 2)
plot_histogram(hue_pepsi, "Histograma de Hue - Pepsi", "blue")

plt.show()
```
